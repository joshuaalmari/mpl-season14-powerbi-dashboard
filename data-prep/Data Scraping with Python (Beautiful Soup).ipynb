# Data Scraping with Python

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from io import StringIO
import os
```

```python
# Fetching URL content
url = 'https://ph-mpl.com/data'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup)
```

```python
# Finding the container holding all the tables
container = soup.find('div', class_='tab-container')
```

```python
# Finding all tables within a container
tables = container.find_all('table')
```

```python
save_folder = r'C:\Users\Joshua Almaril\Mobile Legends - Power BI Project'
os.makedirs(save_folder, exist_ok=True)

table_dfs = []
for i, table in enumerate(tables):
    # Wrap the HTML string in StringIO
    html_string = StringIO(str(table))  # Correctly wrap the HTML string
    df = pd.read_html(html_string)[0]  # Convert HTML table to a DataFrame

    # Construct the file name with MPLS14 and a unique identifier
    file_name = f'MPLS14_{i + 1}.csv'
    file_path = os.path.join(save_folder, file_name)

    # Saving the table into a csv file with the correct path
    df.to_csv(file_path, index=False)

    # Print confirmation
    print(f"Table {i + 1} saved to {file_path}")
```
